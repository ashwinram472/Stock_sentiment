```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import classification_report
import re
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
# Merging both datasets on Date columns
df = pd.merge(df1,df2, on = 'Date')
```

```python
df.shape
```

```python
df.head()
```

```python
# Merge news headlines into one column
headlines = []
for row in range(0, len(df.index)):
    headlines.append(' '.join( str(x) for x in df.iloc[row, 2:27]))
    
# sample
headlines[0]
```

```python
# Cleaning the headlines data
clean_headlines = []

for i in range(0, len(headlines)):
    clean_headlines.append(re.sub("b[(')]" ,'' , headlines[i]))   # remove b'
    clean_headlines[i] = re.sub('b[(")]' , '' , clean_headlines[i]) # remove b"
    clean_headlines[i] = re.sub("\'" , '' , clean_headlines[i]) #removes \
                    
```

```python
# Cleaned Headlines
clean_headlines[0]
df['Combined_News'] = clean_headlines
```

```python
# Look into the mergeed Dataset!
df.head()
```

```python
# Function to get subjectivity
def get_subjectivity(text):
    return TextBlob(text).subjectivity

# Function to get polarity
def get_polarity(text):
    return TextBlob(text).polarity
```

```python
# Create Subjectivity and Polarity Columns
df['Subjectivity'] = df['Combined_News'].apply(get_subjectivity)
df['Polarity'] = df['Combined_News'].apply(get_polarity)
```

```python
# Look into the new columns
df.head()
```

```python
# Function for Sentiment Scores
# Use Sentiment Intensity Analyser:
def getSIA(text):
    sia = SentimentIntensityAnalyzer()
    sentiment = sia.polarity_scores(text)
    return sentiment
```

```python
# Getting the sentiment scores for each day
compound = []
neg  = []
pos  = []
neu  = []
SIA = 0    

for i in range(0, len(df['Combined_News'])):
    SIA = getSIA(df['Combined_News'][i])
    compound.append(SIA['compound'])
    neg.append(SIA['neg'])
    pos.append(SIA['pos'])
    neu.append(SIA['neu'])
```

```python
# Append the sentiment scores to the dataset
df['Compound'] = compound
df['Negative'] = neg
df['Neutral'] = neu
df['Positive'] = pos
```

```python
# Show the merged data
df.head()
```

```python
# List of columns to keep
keep_cols = ['Open', 'High' , 'Low' , 'Volume' , 'Subjectivity' , 'Polarity' , 'Compound' , 'Negative' , 'Neutral' , 'Positive' , 'Label']
```

```python
data = df[keep_cols]
```

```python
#Create Feature Dataset
X  = data
X = np.array(X.drop(['Label'] , 1))

#Create Target Dataset
y = np.array(data['Label'])
```

```python
# Split data into 80 % training and 20 % test
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 7)
```

```python
# Create and train the model
model = LinearDiscriminantAnalysis().fit(x_train, y_train)
```

```python
# Predictions
predict = model.predict(x_test)
predict
```

```python
# Model Metrics
print(classification_report(y_test, predict))
```
